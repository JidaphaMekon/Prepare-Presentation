# ***‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å yahoo finance***

```python
import yfinance as yf
import pandas as pd

# ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô‡∏£‡∏ß‡∏° (‡∏à‡∏≤‡∏Å‡∏™‡∏≠‡∏á‡∏ä‡∏∏‡∏î)
tickers = [

    "AR", "COP", "CTRA", "DVN", "EOG", "EQT", "HAL", "HES", "MPC",
    "MPLX", "OXY", "PR", "PSX", "RRC", "TPL", "VLO"
]

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå Excel ‡πÅ‡∏•‡∏∞‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
with pd.ExcelWriter("energy_largecap_stocks.xlsx", engine='xlsxwriter') as writer:
    for ticker in tickers:
        try:
            df = yf.download(ticker, start="2019-01-01", end="2024-12-31")
            df.to_excel(writer, sheet_name=ticker[:31])  # Sheet name max 31 chars
            print(f"‚úÖ ‡∏î‡∏≤‡∏ß‡∏ô‡πå‡πÇ‡∏´‡∏•‡∏î {ticker} ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à")
        except Exception as e:
            print(f"‚ùå ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• {ticker} ‡πÑ‡∏°‡πà‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: {e}")


```

# ***‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Data Cleaning***

```python
import pandas as pd

# ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå Excel ‡πÄ‡∏î‡∏¥‡∏°
excel = pd.ExcelFile("energy_largecap_stocks.xlsx")

# ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£
columns = ["Date", "Open", "High", "Low", "Close", "Volume"]

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö export
with pd.ExcelWriter("energy_largecap_stocks_cleaned.xlsx") as writer:
    for sheet in excel.sheet_names:
        try:
            # ‡∏≠‡πà‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
            df = pd.read_excel(excel, sheet_name=sheet)

            # ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà 2‚Äì3 (index 0 ‡πÅ‡∏•‡∏∞ 1)
            df.drop(index=[0, 1], inplace=True)

            # ‡∏ï‡∏±‡πâ‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà
            df.columns = columns

            # ‡πÅ‡∏õ‡∏•‡∏á Date ‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡πâ‡∏á‡πÄ‡∏õ‡πá‡∏ô index
            df["Date"] = pd.to_datetime(df["Date"])
            df.set_index("Date", inplace=True)

            # ‡πÅ‡∏õ‡∏•‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô numeric
            df[["Open", "High", "Low", "Close", "Volume"]] = df[["Open", "High", "Low", "Close", "Volume"]].apply(pd.to_numeric, errors="coerce")

            # ‡∏™‡πà‡∏á‡∏≠‡∏≠‡∏Å‡∏•‡∏á Sheet ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà
            df.to_excel(writer, sheet_name=sheet)

        except Exception as e:
            print(f"‚ùå ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡∏´‡∏∏‡πâ‡∏ô {sheet}: {e}")

```

# ***‡∏Ç‡πâ‡∏≤‡∏°‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î ‡∏´‡∏≤‡∏Å‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÑ‡∏ß‡πâ‡πÅ‡∏•‡πâ‡∏ß‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤***

*‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ß‡πà‡∏≤‡∏á‡∏≠‡∏≠‡∏Å‡πÑ‡∏õ*

```python
import pandas as pd
from openpyxl import load_workbook

# üìÇ ‡∏Å‡∏≥‡∏´‡∏ô‡∏î path
file_path = "/content/energy_largecap_stocks_cleaned2019 (2).xlsx"

# üì• ‡πÇ‡∏´‡∏•‡∏î Excel
excel = pd.ExcelFile(file_path)
sheets = {}

# üìä ‡∏≠‡πà‡∏≤‡∏ô‡πÅ‡∏•‡∏∞‡πÅ‡∏õ‡∏•‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
for sheet in excel.sheet_names:
    df = pd.read_excel(excel, sheet_name=sheet)
    df.dropna(how="all", inplace=True)

    # üîÅ ‡πÅ‡∏õ‡∏•‡∏á Date ‡πÄ‡∏õ‡πá‡∏ô datetime ‡πÅ‡∏ö‡∏ö‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÄ‡∏ß‡∏•‡∏≤ (00:00:00)
    if "Date" in df.columns:
        df["Date"] = pd.to_datetime(df["Date"]).dt.normalize()

    sheets[sheet] = df

# üíæ ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Å‡∏•‡∏±‡∏ö‡πÑ‡∏õ
with pd.ExcelWriter(file_path, engine="openpyxl", mode="w") as writer:
    for sheet_name, df in sheets.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

# üé® ‡∏õ‡∏£‡∏±‡∏ö format ‡∏î‡πâ‡∏ß‡∏¢ openpyxl
wb = load_workbook(file_path)

for sheet_name in wb.sheetnames:
    ws = wb[sheet_name]
    header = [cell.value for cell in ws[1]]

    if "Date" in header:
        col_idx = header.index("Date") + 1  # Excel index ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ó‡∏µ‡πà 1
        for row in ws.iter_rows(min_row=2, min_col=col_idx, max_col=col_idx):
            for cell in row:
                cell.number_format = "yyyy-mm-dd"

# ‚úÖ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÅ‡∏õ‡∏•‡∏á
wb.save(file_path)

print("‚úÖ ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á Date ‡πÄ‡∏õ‡πá‡∏ô datetime ‡πÅ‡∏ö‡∏ö‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÄ‡∏ß‡∏•‡∏≤‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß")
```

# ***‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ï‡πà‡∏≠‡πÑ‡∏õ: ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Log Return ‡πÅ‡∏•‡∏∞ Volatility***

```python
import numpy as np

# ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏° list ‡πÄ‡∏Å‡πá‡∏ö summary ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏∏‡πâ‡∏ô
feature_summary = []

for symbol, df in sheets.items():
    # ‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤ Date ‡πÄ‡∏õ‡πá‡∏ô datetime ‡πÅ‡∏•‡∏∞‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ï‡∏≤‡∏°‡∏ß‡∏±‡∏ô
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")

    # ‚úÖ ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Log Return
    df["Log_Return"] = np.log(df["Close"] / df["Close"].shift(1))

    # ‚úÖ ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á Log Return ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ú‡∏±‡∏ô‡∏ú‡∏ß‡∏ô (Volatility)
    mean_return = df["Log_Return"].mean()
    volatility = df["Log_Return"].std()

    # ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß
    feature_summary.append({
        "Symbol": symbol,
        "Mean_Log_Return": mean_return,
        "Volatility": volatility
    })

# ‡∏™‡∏£‡πâ‡∏≤‡∏á DataFrame ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏ï‡πà‡∏≠
features_df = pd.DataFrame(feature_summary)

# ‚úÖ ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
print(features_df)
```

# ***2. ‡∏õ‡∏£‡∏±‡∏ö‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡πâ‡∏ß‡∏¢ StandardScaler***
1. ‡∏õ‡∏£‡∏±‡∏ö‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Standardization) ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏°‡∏µ‡∏™‡πÄ‡∏Å‡∏•‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô‡∏Å‡πà‡∏≠‡∏ô‡∏ô‡∏≥‡πÑ‡∏õ‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏•

\\
‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå


*   ‡∏Ñ‡πà‡∏≤ ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (Mean) ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞ feature ‚âà 0
*   ‡∏Ñ‡πà‡∏≤ ‡∏™‡πà‡∏ß‡∏ô‡πÄ‡∏ö‡∏µ‡πà‡∏¢‡∏á‡πÄ‡∏ö‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô (Std) ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞ feature ‚âà 1



2.‡πÉ‡∏´‡πâ‡∏ó‡∏∏‡∏Å‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏°‡∏µ ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á Euclidean (‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡πÉ‡∏ô K-Means)

```python
from sklearn.decomposition import PCA
import pandas as pd

# ‡∏™‡∏°‡∏°‡∏ï‡∏¥ X_scaled ‡πÄ‡∏õ‡πá‡∏ô DataFrame ‡∏´‡∏•‡∏±‡∏á StandardScaler

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏°‡πÇ‡∏û‡πÄ‡∏ô‡∏ô‡∏ï‡πå‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏Å‡πá‡∏ö (‡πÄ‡∏ä‡πà‡∏ô 2)
pca = PCA(n_components=2)

# Fit ‡πÅ‡∏•‡∏∞ Transform ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
X_pca = pca.fit_transform(X_scaled)

# ‡πÅ‡∏õ‡∏•‡∏á‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏õ‡πá‡∏ô DataFrame ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π‡∏á‡πà‡∏≤‡∏¢ ‡πÜ
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

# ‡∏î‡∏π‡∏Ñ‡πà‡∏≤ Variance Ratio ‡∏ó‡∏µ‡πà‡πÅ‡∏ï‡πà‡∏•‡∏∞ PC ‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢‡πÑ‡∏î‡πâ
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)

# ‡∏î‡∏π DataFrame PCA ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á
print(df_pca.head())
```

# ***‡πÉ‡∏ä‡πâ Silhouette Score ‡∏ß‡∏±‡∏î‡∏Ñ‡∏∏‡∏ì‡∏†‡∏≤‡∏û‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Å‡∏•‡∏∏‡πà‡∏°‡πÉ‡∏ô‡∏ä‡πà‡∏ß‡∏á K=2 ‡∏ñ‡∏∂‡∏á 10***
(‡∏ä‡πà‡∏ß‡∏¢‡∏´‡∏≤‡∏ß‡πà‡∏≤ K ‡πÑ‡∏´‡∏ô‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# ‡πÉ‡∏ä‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å PCA
scaled_features = df_pca[['PC1', 'PC2']]  # ‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà‡∏à‡∏∞‡πÉ‡∏ä‡πâ‡∏à‡∏≤‡∏Å StandardScaler ‡πÇ‡∏î‡∏¢‡∏ï‡∏£‡∏á

sil_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    labels = kmeans.labels_
    score = silhouette_score(scaled_features, labels)
    sil_scores.append(score)

# ‡∏´‡∏≤‡∏Ñ‡πà‡∏≤ K ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
best_k = np.argmax(sil_scores) + 2
print(f"Best K: {best_k}")
```

 ‡∏ß‡∏¥‡∏ò‡∏µ‡∏Å‡∏≤‡∏£‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏ó‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏•‡∏±‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå
‡∏´‡∏•‡∏±‡∏Å‡∏Å‡∏≤‡∏£:


‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà ‚Äú‡πÉ‡∏Å‡∏•‡πâ‡∏®‡∏π‡∏ô‡∏¢‡πå‡∏Å‡∏•‡∏≤‡∏á‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏•‡∏±‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå‚Äù ‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡πÉ‡∏ô‡πÄ‡∏ä‡∏¥‡∏á‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á PCA (PC1, PC2)

```python
import numpy as np

representatives = []

# ‡∏ß‡∏ô‡∏ú‡πà‡∏≤‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏•‡∏±‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå
for cluster_id in df_pca_with_name['cluster'].unique():
    cluster_data = df_pca_with_name[df_pca_with_name['cluster'] == cluster_id]

    # ‡∏î‡∏∂‡∏á‡∏à‡∏∏‡∏î‡∏®‡∏π‡∏ô‡∏¢‡πå‡∏Å‡∏•‡∏≤‡∏á‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏•‡∏±‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå
    center = cluster_data[['PC1', 'PC2']].mean().values

    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏£‡∏∞‡∏¢‡∏∞‡∏´‡πà‡∏≤‡∏á‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏à‡∏∏‡∏î‡πÉ‡∏ô‡∏Ñ‡∏•‡∏±‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå‡∏à‡∏≤‡∏Å‡∏à‡∏∏‡∏î‡∏®‡∏π‡∏ô‡∏¢‡πå‡∏Å‡∏•‡∏≤‡∏á
    distances = np.linalg.norm(cluster_data[['PC1', 'PC2']].values - center, axis=1)

    # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà‡πÉ‡∏Å‡∏•‡πâ‡∏®‡∏π‡∏ô‡∏¢‡πå‡∏Å‡∏•‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
    min_idx = distances.argmin()
    representative = cluster_data.iloc[min_idx]['Symbol']

    representatives.append((cluster_id, representative))

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏ó‡∏ô
for cluster_id, symbol in representatives:
    print(f"Cluster {cluster_id} representative stock: {symbol}")
```

```python
import numpy as np

representatives_top2 = []

for cluster_id in df_pca_with_name['cluster'].unique():
    cluster_data = df_pca_with_name[df_pca_with_name['cluster'] == cluster_id].copy()
    center = cluster_data[['PC1', 'PC2']].mean().values
    cluster_data['distance'] = np.linalg.norm(cluster_data[['PC1', 'PC2']].values - center, axis=1)
    cluster_data_sorted = cluster_data.sort_values('distance').reset_index(drop=True)

    top1 = cluster_data_sorted.loc[0, 'Symbol']
    top1_dist = cluster_data_sorted.loc[0, 'distance']

    # ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏≠‡∏±‡∏ô‡∏î‡∏±‡∏ö 2 ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
    if len(cluster_data_sorted) > 1:
        top2 = cluster_data_sorted.loc[1, 'Symbol']
        top2_dist = cluster_data_sorted.loc[1, 'distance']
    else:
        top2 = None
        top2_dist = None

    representatives_top2.append({
        'Cluster': cluster_id,
        'Top1': top1,
        'Top1_Distance': top1_dist,
        'Top2': top2,
        'Top2_Distance': top2_dist
    })

df_top2 = pd.DataFrame(representatives_top2)
print(df_top2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
colors = ['red', 'blue', 'green']

# ‡∏û‡∏•‡πá‡∏≠‡∏ï‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏•‡∏±‡∏™‡πÄ‡∏ï‡∏≠‡∏£‡πå
for i in df_pca_with_name['cluster'].unique():
    cluster_data = df_pca_with_name[df_pca_with_name['cluster'] == i]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], c=colors[i], alpha=0.5, label=f'Cluster {i}')

# ‡∏û‡∏•‡πá‡∏≠‡∏ï Top1 ‡πÅ‡∏•‡∏∞ Top2
for _, row in df_top2.iterrows():
    # Top1
    top1_row = df_pca_with_name[df_pca_with_name['Symbol'] == row['Top1']]
    plt.scatter(top1_row['PC1'], top1_row['PC2'], c='black', s=150, marker='*', edgecolor='white', label=f"Top1 {row['Top1']}")
    plt.text(top1_row['PC1'].values[0]+0.02, top1_row['PC2'].values[0], row['Top1'], fontsize=10, weight='bold')

    # Top2 (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
    if row['Top2'] is not None:
        top2_row = df_pca_with_name[df_pca_with_name['Symbol'] == row['Top2']]
        plt.scatter(top2_row['PC1'], top2_row['PC2'], c='orange', s=120, marker='P', edgecolor='white', label=f"Top2 {row['Top2']}")
        plt.text(top2_row['PC1'].values[0]+0.02, top2_row['PC2'].values[0], row['Top2'], fontsize=10, color='orange')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clusters with Top1 (black star) and Top2 (orange pentagon)')
plt.legend()
plt.grid(True)
plt.show()
```
