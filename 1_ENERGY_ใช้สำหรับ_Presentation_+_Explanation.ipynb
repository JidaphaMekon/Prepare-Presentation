# ***ดึงข้อมูลจาก yahoo finance***

```python
import yfinance as yf
import pandas as pd

# รายชื่อหุ้นรวม (จากสองชุด)
tickers = [

    "AR", "COP", "CTRA", "DVN", "EOG", "EQT", "HAL", "HES", "MPC",
    "MPLX", "OXY", "PR", "PSX", "RRC", "TPL", "VLO"
]

# สร้างไฟล์ Excel และดึงข้อมูล
with pd.ExcelWriter("energy_largecap_stocks.xlsx", engine='xlsxwriter') as writer:
    for ticker in tickers:
        try:
            df = yf.download(ticker, start="2019-01-01", end="2024-12-31")
            df.to_excel(writer, sheet_name=ticker[:31])  # Sheet name max 31 chars
            print(f"✅ ดาวน์โหลด {ticker} สำเร็จ")
        except Exception as e:
            print(f"❌ ดึงข้อมูล {ticker} ไม่สำเร็จ: {e}")


```

# ***วิธีทำความสะอาดข้อมูล Data Cleaning***

```python
import pandas as pd

# โหลดไฟล์ Excel เดิม
excel = pd.ExcelFile("energy_largecap_stocks.xlsx")

# เตรียมชื่อคอลัมน์ที่ต้องการ
columns = ["Date", "Open", "High", "Low", "Close", "Volume"]

# สร้างไฟล์ใหม่สำหรับ export
with pd.ExcelWriter("energy_largecap_stocks_cleaned.xlsx") as writer:
    for sheet in excel.sheet_names:
        try:
            # อ่านข้อมูล
            df = pd.read_excel(excel, sheet_name=sheet)

            # ลบแถวที่ 2–3 (index 0 และ 1)
            df.drop(index=[0, 1], inplace=True)

            # ตั้งชื่อคอลัมน์ใหม่
            df.columns = columns

            # แปลง Date และตั้งเป็น index
            df["Date"] = pd.to_datetime(df["Date"])
            df.set_index("Date", inplace=True)

            # แปลงข้อมูลให้เป็น numeric
            df[["Open", "High", "Low", "Close", "Volume"]] = df[["Open", "High", "Low", "Close", "Volume"]].apply(pd.to_numeric, errors="coerce")

            # ส่งออกลง Sheet เดียวกันในไฟล์ใหม่
            df.to_excel(writer, sheet_name=sheet)

        except Exception as e:
            print(f"❌ ไม่สามารถประมวลผลหุ้น {sheet}: {e}")

```

# ***ข้ามขั้นตอนการดึงข้อมูลและการทำความสะอาด หากเตรียมข้อมูลไว้แล้วก่อนหน้า***

*ดึงข้อมูลเพื่อปรับข้อมูล ลบแถวที่ว่างออกไป*

```python
import pandas as pd
from openpyxl import load_workbook

# 📂 กำหนด path
file_path = "/content/energy_largecap_stocks_cleaned2019 (2).xlsx"

# 📥 โหลด Excel
excel = pd.ExcelFile(file_path)
sheets = {}

# 📊 อ่านและแปลงข้อมูล
for sheet in excel.sheet_names:
    df = pd.read_excel(excel, sheet_name=sheet)
    df.dropna(how="all", inplace=True)

    # 🔁 แปลง Date เป็น datetime แบบไม่มีเวลา (00:00:00)
    if "Date" in df.columns:
        df["Date"] = pd.to_datetime(df["Date"]).dt.normalize()

    sheets[sheet] = df

# 💾 เขียนข้อมูลกลับไป
with pd.ExcelWriter(file_path, engine="openpyxl", mode="w") as writer:
    for sheet_name, df in sheets.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

# 🎨 ปรับ format ด้วย openpyxl
wb = load_workbook(file_path)

for sheet_name in wb.sheetnames:
    ws = wb[sheet_name]
    header = [cell.value for cell in ws[1]]

    if "Date" in header:
        col_idx = header.index("Date") + 1  # Excel index เริ่มที่ 1
        for row in ws.iter_rows(min_row=2, min_col=col_idx, max_col=col_idx):
            for cell in row:
                cell.number_format = "yyyy-mm-dd"

# ✅ บันทึกการเปลี่ยนแปลง
wb.save(file_path)

print("✅ เขียนและแสดง Date เป็น datetime แบบไม่มีเวลาเรียบร้อยแล้ว")
```

# ***ขั้นตอนต่อไป: คำนวณ Log Return และ Volatility***

```python
import numpy as np

# เตรียม list เก็บ summary ของแต่ละหุ้น
feature_summary = []

for symbol, df in sheets.items():
    # ให้แน่ใจว่า Date เป็น datetime และเรียงตามวัน
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")

    # ✅ คำนวณ Log Return
    df["Log_Return"] = np.log(df["Close"] / df["Close"].shift(1))

    # ✅ คำนวณค่าเฉลี่ยของ Log Return และความผันผวน (Volatility)
    mean_return = df["Log_Return"].mean()
    volatility = df["Log_Return"].std()

    # เก็บข้อมูลของหุ้นแต่ละตัว
    feature_summary.append({
        "Symbol": symbol,
        "Mean_Log_Return": mean_return,
        "Volatility": volatility
    })

# สร้าง DataFrame สำหรับใช้งานต่อ
features_df = pd.DataFrame(feature_summary)

# ✅ แสดงผล
print(features_df)
```

# ***PCA***

```python
from sklearn.decomposition import PCA
import pandas as pd

# สมมติ X_scaled เป็น DataFrame หลัง StandardScaler

# กำหนดจำนวนคอมโพเนนต์ที่ต้องการเก็บ (เช่น 2)
pca = PCA(n_components=2)

# Fit และ Transform ข้อมูล
X_pca = pca.fit_transform(X_scaled)

# แปลงกลับเป็น DataFrame เพื่อดูง่าย ๆ
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

# ดูค่า Variance Ratio ที่แต่ละ PC อธิบายได้
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)

# ดู DataFrame PCA ตัวอย่าง
print(df_pca.head())
```

# ***ใช้ Silhouette Score วัดคุณภาพการแบ่งกลุ่มในช่วง K=2 ถึง 10***
(ช่วยหาว่า K ไหนเหมาะสมที่สุด)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# ใช้ข้อมูลจาก PCA
scaled_features = df_pca[['PC1', 'PC2']]  # แทนที่จะใช้จาก StandardScaler โดยตรง

sil_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    labels = kmeans.labels_
    score = silhouette_score(scaled_features, labels)
    sil_scores.append(score)

# หาค่า K ที่ดีที่สุด
best_k = np.argmax(sil_scores) + 2
print(f"Best K: {best_k}")
```

# ***Elbow Method (Plot inertia) ***
(ช่วยตรวจสอบว่า K ที่ได้จาก Silhouette Score สอดคล้องกับจุด elbow หรือไม่)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

# ***(Silhouette อีกชุด)***

Silhouette Score อีกชุดกับช่วง K=3 ถึง 10 เพื่อเช็คความเปลี่ยนแปลง
(ถ้าอยากลองช่วง K ที่แตกต่างจากชุดแรก)



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sil_scores = []
for k in range(3, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    labels = kmeans.labels_
    score = silhouette_score(scaled_features, labels)
    sil_scores.append(score)

best_k = np.argmax(sil_scores) + 3 # เนื่องจากเริ่มต้นที่ K=2
print(f"Best K: {best_k}")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# กำหนดช่วงของ Best K ที่ต้องการลองรัน
best_k_values = [2, 3, 4, 5, 6, 7]  # แก้ไขตามค่า Best K ที่สนใจ

colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

plt.figure(figsize=(15, 10))

for idx, k in enumerate(best_k_values, start=1):
    # สร้างโมเดล KMeans
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)

    # เพิ่มคอลัมน์ cluster ชั่วคราว (ไม่เปลี่ยน df_pca จริงๆ)
    clusters = kmeans.labels_

    # วาดกราฟ subplot แยกแต่ละ K
    plt.subplot(2, 3, idx)  # จัด subplot 2 แถว 3 คอลัมน์

    for i in range(k):
        cluster_data = scaled_features[clusters == i]
        plt.scatter(cluster_data['PC1'], cluster_data['PC2'],
                    s=30, c=colors[i], label=f'Cluster {i}')
    plt.title(f'K-Means with K={k}')
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

representatives = []

# วนผ่านแต่ละคลัสเตอร์
for cluster_id in df_pca_with_name['cluster'].unique():
    cluster_data = df_pca_with_name[df_pca_with_name['cluster'] == cluster_id]

    # ดึงจุดศูนย์กลางของคลัสเตอร์
    center = cluster_data[['PC1', 'PC2']].mean().values

    # คำนวณระยะห่างของแต่ละจุดในคลัสเตอร์จากจุดศูนย์กลาง
    distances = np.linalg.norm(cluster_data[['PC1', 'PC2']].values - center, axis=1)

    # เลือกจุดที่ใกล้ศูนย์กลางที่สุด
    min_idx = distances.argmin()
    representative = cluster_data.iloc[min_idx]['Symbol']

    representatives.append((cluster_id, representative))

# แสดงผลหุ้นตัวแทน
for cluster_id, symbol in representatives:
    print(f"Cluster {cluster_id} representative stock: {symbol}")
```

```python
import numpy as np

representatives_top2 = []

for cluster_id in df_pca_with_name['cluster'].unique():
    cluster_data = df_pca_with_name[df_pca_with_name['cluster'] == cluster_id].copy()
    center = cluster_data[['PC1', 'PC2']].mean().values
    cluster_data['distance'] = np.linalg.norm(cluster_data[['PC1', 'PC2']].values - center, axis=1)
    cluster_data_sorted = cluster_data.sort_values('distance').reset_index(drop=True)

    top1 = cluster_data_sorted.loc[0, 'Symbol']
    top1_dist = cluster_data_sorted.loc[0, 'distance']

    # เช็คว่ามีอันดับ 2 หรือไม่
    if len(cluster_data_sorted) > 1:
        top2 = cluster_data_sorted.loc[1, 'Symbol']
        top2_dist = cluster_data_sorted.loc[1, 'distance']
    else:
        top2 = None
        top2_dist = None

    representatives_top2.append({
        'Cluster': cluster_id,
        'Top1': top1,
        'Top1_Distance': top1_dist,
        'Top2': top2,
        'Top2_Distance': top2_dist
    })

df_top2 = pd.DataFrame(representatives_top2)
print(df_top2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
colors = ['red', 'blue', 'green']

# พล็อตข้อมูลแต่ละคลัสเตอร์
for i in df_pca_with_name['cluster'].unique():
    cluster_data = df_pca_with_name[df_pca_with_name['cluster'] == i]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], c=colors[i], alpha=0.5, label=f'Cluster {i}')

# พล็อต Top1 และ Top2
for _, row in df_top2.iterrows():
    # Top1
    top1_row = df_pca_with_name[df_pca_with_name['Symbol'] == row['Top1']]
    plt.scatter(top1_row['PC1'], top1_row['PC2'], c='black', s=150, marker='*', edgecolor='white', label=f"Top1 {row['Top1']}")
    plt.text(top1_row['PC1'].values[0]+0.02, top1_row['PC2'].values[0], row['Top1'], fontsize=10, weight='bold')

    # Top2 (ถ้ามี)
    if row['Top2'] is not None:
        top2_row = df_pca_with_name[df_pca_with_name['Symbol'] == row['Top2']]
        plt.scatter(top2_row['PC1'], top2_row['PC2'], c='orange', s=120, marker='P', edgecolor='white', label=f"Top2 {row['Top2']}")
        plt.text(top2_row['PC1'].values[0]+0.02, top2_row['PC2'].values[0], row['Top2'], fontsize=10, color='orange')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clusters with Top1 (black star) and Top2 (orange pentagon)')
plt.legend()
plt.grid(True)
plt.show()
```
